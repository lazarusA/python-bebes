```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logi_sol(t, x0, r):
    return 1/(1 + (1/x0 - 1) * np.exp(-r * t))
```

```python
t = np.linspace(0,10)
for r in np.arange(-1, 1,.1):
    plt.plot(t, logi_sol(t, 0.5, r))
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.show()
```

### Actividad
ponerle etiquetas a cada curva

```python
t = np.linspace(0,10)
for r in np.arange(-1, 1,.1):
    plt.plot(t, logi_sol(t, 0.5, r), label = r)
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.legend(bbox_to_anchor=(1.09,1), prop={'size': 10})  
plt.show()
```

```python
t = np.linspace(0,10)
for r in np.arange(-1, 1,.25):
    plt.plot(t, logi_sol(t, 0.5, r), label = '%s' % r)
plt.xlabel('$t$', fontsize = 18)
plt.ylabel(r'$x$', fontsize = 18)
plt.legend(bbox_to_anchor=(1.25,.9), prop={'size': 10})  
plt.show()
```

### Mapa logstico

```python
def mapa_logistico(r, x):
    return r * x * (1 - x)
```

```python
# Para mil valores de r entre 2.5 y 4.0
n = 1000
r = np.linspace(3.81, 3.87, n)
```

```python
# Hacemos 1000 iteraciones y nos quedamos con las ultimas 100
iterar = 1000
ultimos = 500
```

```python
# La misma condicin inicial para todos los casos. 
x = 1e-5 * np.ones(n)
```

```python
plt.figure(figsize=(7, 4.5))
for i in range(iterar):
    x = mapa_logistico(r, x)
    if i >= (iterar - ultimos):
        plt.plot(r, x, ',k', alpha=.2)
plt.xlim(3.8, 3.87)
plt.ylim(-.1, 1.1)
plt.title("Diagrama de bifurcacin")
plt.xlabel('$r$', fontsize=18)
plt.ylabel('$x$', fontsize=18)
plt.show()
```
